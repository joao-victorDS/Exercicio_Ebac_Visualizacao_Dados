#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
# Filtrar para categorizar variavel resposta
data_sus.loc[(data_sus["APGAR5"] <= 3) & (data_sus["APGAR5"] >= 0), "cat_APGAR5"] = "asfixia grave"
data_sus.loc[(data_sus["APGAR5"] <= 5) & (data_sus["APGAR5"] >= 4), "cat_APGAR5"] = "asfixia moderada"
data_sus.loc[(data_sus["APGAR5"] <= 7) & (data_sus["APGAR5"] >= 6), "cat_APGAR5"] = "asfixia leve"
data_sus.loc[(data_sus["APGAR5"] <=10) & (data_sus["APGAR5"] >= 8), "cat_APGAR5"] = "normal"

data_sus.head(10)
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Ajustar tamanho da figura e ordenar os ticks
plt.rc("figure", figsize=(10,7))
order = ["Menos de 22 semanas", "22 a 27 semanas", "28 a 31 semanas", 
         "32 a 36 semanas", "37 a 41 semanas", "42 semanas e mais"]

# Plotar gráfico da variavel resposta em relação ao tempo de gestacao
sns.boxplot(data=data_sus, y="GESTACAO", x="APGAR5", hue="SEXO", order=order)

# Mudanças esteticas
plt.legend(loc="lower left")
plt.title("APGAR 5 em Relação ao Tempo de Gestação", size=14)
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Cruzar dados de tempo de gestacao e apgar5
freq = pd.crosstab(data_sus["GESTACAO"], data_sus["APGAR5"], )
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Somar as linhas da tabela de cruzamento anterior
sum_freq = freq.sum(axis=1)

# Dividir os valores pelo resultado da soma
pct_freq = round(freq.div(sum_freq, axis=0) * 100, 2)
pct_freq
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas.
```python
# Filtrar para categorizar a variavel peso
data_sus.loc[(data_sus["PESO"] < 2500) & (data_sus["SEXO"] == "Masculino"), "cat_PESO"] = "Abaixo do Peso"
data_sus.loc[(data_sus["PESO"] < 2400) & (data_sus["SEXO"] == "Feminino"), "cat_PESO"] = "Abaixo do Peso"

data_sus.loc[(data_sus["PESO"] >= 2500) & (data_sus["PESO"] <= 4400) & (data_sus["SEXO"] == "Masculino"), "cat_PESO"] = "Peso Normal"
data_sus.loc[(data_sus["PESO"] >= 2400) & (data_sus["PESO"] <= 4200) & (data_sus["SEXO"] == "Feminino"), "cat_PESO"] = "Peso Normal"

data_sus.loc[(data_sus["PESO"] > 4400) & (data_sus["SEXO"] == "Masculino"), "cat_PESO"] = "Acima do Peso"
data_sus.loc[(data_sus["PESO"] > 4200) & (data_sus["SEXO"] == "Feminino"), "cat_PESO"] = "Acima do Peso"

data_sus.cat_PESO.value_counts()
```

```python
# Linhas aleatorias 
data_sus[["PESO", "cat_PESO", "APGAR5", "cat_APGAR5"]].sample(10)
```

```python
# Cruzar as variaveis peso e apgar5
freq = pd.crosstab(data_sus["cat_PESO"], data_sus["cat_APGAR5"])

# Somar as linhas da tabela de dados cruzados
sum_freq = freq.sum(axis=1)

# Dividir os dados da tabela pelo resultado da soma das suas rescpectivas linhas
pct_freq = round(freq.div(sum_freq, axis=0) * 100, 2)
pct_freq
```

```python
# Plotar grafico da tabela de dados cruzados 
pct_freq.plot.bar()

# Modificações esteticas 
plt.title("Peso x Apgar 5", size=16)
plt.ylabel("Porcentagem", size=12)
plt.xlabel("Categoria do Peso (RN)", size=12)
plt.show()
```

```python
# Redefinir tamanho da figura
plt.rc("figure", figsize=(12, 7))

# Plotar grafico da variavel apgar5 em relação ao peso
sns.lineplot(data=data_sus, x="APGAR5", y="PESO", errorbar=("ci", False), hue="SEXO", marker="o")

# Modificaçoes esteticas
plt.title("Peso X Apgar 5", size=16)
```

### Insights
* **Existe uma tendência de asfixia para recem nascidos de menor peso**
* **Bebês com peso normal e acima do peso, no geral, tem uma taxa proxima de 8% a mais para APGAR5 normal**
* **A taxa de asfixia leve é 6,25 vezes maior para bebês abaixo do peso do que para acima do peso, e 6.39 vezes para bebês com peso normal**
* **Bebês do sexo feminino com 1000g a 1500g são propicias a asfixia grave**

### Conclusão
* **O peso tem sim uma relevância no nivel de APGAR5 do recem nascido. Cerca de 8% dos bebês abaixo do peso tiveram alguma complicação com a respiração, comparado a, aproximadamente 1,5% dos bebês acima do peso e com peso normal. Isso significa dizer que bebês mais pesados  tendem a ter menos complicações com a respiração**
